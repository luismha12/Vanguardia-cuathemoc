```python
# Análisis del Conjunto de Datos Iris

# Importar las librerías necesarias
from sklearn.datasets import fetch_openml  # type: ignore
import pandas as pd  # type: ignore

# Descargar la base de datos "Iris" desde OpenML, con parser='auto' para evitar la advertencia
dataset = fetch_openml(name='iris', version=1, as_frame=True, parser='auto')

# Crear un DataFrame a partir de los datos descargados
df = dataset.frame

# Mostrar las primeras filas del DataFrame
print("Primeras filas del dataset 'Iris':")
display(df.head())  # Para mostrar en formato de tabla en Jupyter

# Descripción general de las variables
print("\nDescripción de las variables del dataset:")
df.info()  # Proporciona información sobre los tipos de variables y valores faltantes

# Identificación del tipo de variable (numérica discreta, continua, categórica)
print("\nIdentificación de los tipos de variables:")
for column in df.columns:
    dtype = df[column].dtype
    if pd.api.types.is_numeric_dtype(dtype):
        if pd.api.types.is_integer_dtype(dtype):
            var_type = "Numérica Discreta"
        else:
            var_type = "Numérica Continua"
    else:
        var_type = "Categórica"
    
    print(f"Variable '{column}': {var_type}")
```
