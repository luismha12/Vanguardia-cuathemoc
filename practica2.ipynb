```python
# **Análisis** del Conjunto de Datos Iris con Visualización

# Este script en Python carga y analiza el conjunto de datos Iris utilizando las bibliotecas pandas, matplotlib y seaborn.

# Importar las librerías necesarias
from sklearn.datasets import load_iris  # type: ignore
import pandas as pd  # type: ignore
import matplotlib.pyplot as plt  # type: ignore
import seaborn as sns  # type: ignore

# Configuración de estilo de gráficos
sns.set(style="whitegrid")

# Descargar el conjunto de datos Iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# 1. Parámetros estadísticos de todas las variables
print("Parámetros estadísticos de todas las variables:")
print(df.describe())

# 2. Crear gráficas

# Gráfico 1: Gráfico de dispersión (scatter plot) de Sepal Length vs Sepal Width
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', palette='viridis', data=df)
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species', labels=iris.target_names)
plt.show()

# Gráfico 2: Box plot de las características
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.iloc[:, :-1])  # Excluimos la columna 'target' para el boxplot
plt.title('Box Plot de las Características')
plt.xticks(rotation=45)
plt.show()

# Gráfico 3: Histogramas de las características
df.iloc[:, :-1].hist(bins=15, figsize=(15, 6), layout=(1, 4), color='skyblue', edgecolor='black')
plt.suptitle('Histogramas de las Características')
plt.show()

# Gráfico 4: Gráfico de pares (pairplot) para observar relaciones
sns.pairplot(df, hue='target', palette='viridis', markers=["o", "s", "D"], diag_kind='kde')
plt.suptitle('Pairplot de las Características', y=1.02)
plt.show()
```

```python
# Análisis del Conjunto de Datos Iris

# Importar las librerías necesarias
from sklearn.datasets import load_iris  # type: ignore
import pandas as pd  # type: ignore

# Descargar el conjunto de datos Iris
iris = load_iris()

# Crear un DataFrame de pandas
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Mostrar las primeras filas del conjunto de datos
print("Primeras filas del dataset Iris:")
print(df.head())

# Describir las variables y su tipo
print("\nDescripción de las variables del dataset:")
print(df.info())  # Esto te dará una descripción general de cada variable

# Identificación del tipo de variable
print("\nIdentificación de los tipos de variables:")
for column in df.columns:
    dtype = df[column].dtype
    if pd.api.types.is_numeric_dtype(dtype):
        if pd.api.types.is_integer_dtype(dtype):
            var_type = "Numérica Discreta"
        else:
            var_type = "Numérica Continua"
    else:
        var_type = "Categórica"
    
    print(f"Variable '{column}': {var_type}")
```
